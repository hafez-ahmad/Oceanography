```
import pandas as ps
import matplotlib.pyplot as plt
```

```
import statsmodels.api as sm
```

```
data=sm.datasets.co2.load_pandas()
```

```
co2=data.data
```

```
co2.head()
```

```
co2.index
```

```
co2
```

```
y=co2['co2'].resample('MS').mean()
```

```
y
```

```
y['2001']
```

```
y.isnull().sum()
```

```
y=y.fillna(y.bfill())
```

```
y.isnull().sum()
```

```
y.plot(figsize=(10,5))
```

```
decomposition=sm.tsa.seasonal_decompose(y,model='additive')
decomposition.plot()
```
